```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
sample1 = np.random.normal(1, 5, 1000)                    
```

```python
def outliers_funct(sample1,num1):
        sample_copy = sample1.copy()
        for j in range(num1):
            sample_copy[j] = np.random.randint(450,650)
                
        return np.mean(sample_copy),np.std(sample_copy),np.median(sample_copy),np.percentile(sample_copy,75)-np.percentile(sample_copy,25)
```

```python
all_mean=[]
all_median = []
all_std = []
all_iqr = []
```

```python
for i in range(1,51):
    mean1,std1,median1,iqr1 = outliers_funct(sample1,i)
    
    all_mean.append(mean1)
    all_median.append(median1)
    all_iqr.append(iqr1)
    all_std.append(std1)
    if(i % 5 == 0):
        print(f"Number of outliers: {i}, Mean: {mean1}, Std: {std1}, Median: {median1}, IQR: {iqr1}")
        

plt.plot(all_mean, label='Mean', color='blue', marker='o')
plt.plot(all_median, label='Median', color='red', marker='o')
plt.plot(all_std, label='Std', color='green', marker='o')
plt.plot(all_iqr, label='IQR', color='orange', marker='o')

plt.xlabel('Number of Outliers')
plt.ylabel('Value')
plt.title('Effect of Outliers on All the  Data')
plt.legend()
plt.grid()
plt.show()
```
